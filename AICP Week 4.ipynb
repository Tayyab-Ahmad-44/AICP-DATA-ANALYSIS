# Exploratory Data Analysis Internship Task Week 4

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

## Task 1
#### Read menu.csv file

```python
df = pd.read_csv("menu.csv")
```

#### Selecting relevent attributes 
['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily
Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 'Iron (% Daily Value)']

```python
selected_attributes = ['Calories', 'Total Fat', 'Carbohydrates',
                       'Dietary Fiber', 'Sugars', 'Protein',
                       'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
                       'Calcium (% Daily Value)', 'Iron (% Daily Value)']
```

# Task 2

#### Statistical Summary

```python
df.describe().round(2)
```

#### Maximum Values

```python
df[selected_attributes].max()
```

# Task 3
#### Corelation Matrix

```python
correlation_matrix = df[selected_attributes].corr()

# Generate a heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix: McDonald's Menu Nutritional Values")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.show()
```

# Task 4
#### Boxplot for Calories vs Category

```python
sns.boxplot(
    x = "Category",
    y = "Calories",
    showmeans=True,  # Display means as markers
    data=df
)

# Customize the plot
plt.title("Distribution of Calories by Category")
plt.xlabel("Category")
plt.ylabel("Calories")
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better readability
plt.show()
```

# Task 5

```python

highest_calories_item = df.loc[df['Calories'].idxmax()]
highest_total_fat_item = df.loc[df['Total Fat'].idxmax()]
highest_carbohydrates_item = df.loc[df['Carbohydrates'].idxmax()]
highest_dietary_fiber_item = df.loc[df['Dietary Fiber'].idxmax()]
highest_sugars_item = df.loc[df['Sugars'].idxmax()]
highest_protein_item = df.loc[df['Protein'].idxmax()]
highest_vitamin_a_item = df.loc[df['Vitamin A (% Daily Value)'].idxmax()]
highest_vitamin_c_item = df.loc[df['Vitamin C (% Daily Value)'].idxmax()]
highest_calcium_item = df.loc[df['Calcium (% Daily Value)'].idxmax()]
highest_iron_item = df.loc[df['Iron (% Daily Value)'].idxmax()]

# Display
print("Item with highest quantity for each attribute:")
print("Calories:", highest_calories_item['Item'], f"({highest_calories_item['Calories']:.0f})")
print("Total Fat:", highest_total_fat_item['Item'], f"({highest_total_fat_item['Total Fat']:.0f})")
print("Carbohydrates:", highest_carbohydrates_item['Item'], f"({highest_carbohydrates_item['Carbohydrates']:.0f})")
print("Dietary Fiber:", highest_dietary_fiber_item['Item'], f"({highest_dietary_fiber_item['Dietary Fiber']:.0f})")
print("Sugars:", highest_sugars_item['Item'], f"({highest_sugars_item['Sugars']:.0f})")
print("Protein:", highest_protein_item['Item'], f"({highest_protein_item['Protein']:.0f})")
print("Vitamin A (% Daily Value):", highest_vitamin_a_item['Item'], f"({highest_vitamin_a_item['Vitamin A (% Daily Value)']:.0f})")
print("Vitamin C (% Daily Value):", highest_vitamin_c_item['Item'], f"({highest_vitamin_c_item['Vitamin C (% Daily Value)']:.0f})")
print("Calcium (% Daily Value):", highest_calcium_item['Item'], f"({highest_calcium_item['Calcium (% Daily Value)']:.0f})")
print("Iron (% Daily Value):", highest_iron_item['Item'], f"({highest_iron_item['Iron (% Daily Value)']:.0f})")
```

# Task 6
#### Stripplot for each category against specified attributes

```python
attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein',
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)',
              'Iron (% Daily Value)']

for attribute in attributes:
    plt.figure(figsize=(10, 6))
    sns.stripplot(x='Category', y=attribute, data=df)
    plt.title(f'Stripplot of {attribute} by Category')
    plt.xticks(rotation=45)
    plt.show()
```

# Task 7
#### Horizontal bar graph for items in each category against the calories

```python
categories = df['Category'].unique()
for category in categories:
    category_df = df[df['Category'] == category]
    plt.barh(category_df['Item'], category_df['Calories'])
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.title(f'Calories in {category}')
    plt.gca().invert_yaxis()  # Arrange bars from highest to lowest calories
    plt.show()

# Analyze observations based on the patterns and trends in the graphs.
```
